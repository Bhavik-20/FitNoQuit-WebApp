```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('exercise_dataset.csv')
df.drop(['Calories per kg'], axis = 1, inplace = True)
df
```

```python
df['CPK'] = 1
for i in range(0, len(df)):
    df['CPK'][i] = (df['130 lb'][i] / 59.1 + df['155 lb'][i] / 70.45 + df['180 lb'][i] / 81.81 + df['205 lb'][i] / 93.18) / 4
df
```

```python
calories = 200
time = [0.5, 0.75, 1] #30 mins, 45 mins, 1hr
w = 70
l = []
for i in range(0, len(df)):
    cpk = df['CPK'][i] * w
    for t in time:
        if cpk * t <= 220 and cpk * t >= 180:
            l.append((df['Activity, Exercise or Sport (1 hour)'][i], t, cpk * t))
l
```

```python
import random
import math
import statistics
```

```python
init_centroids = [df.iloc[random.randint(0,86)], df.iloc[random.randint(0,86)], df.iloc[random.randint(0,86)]]
clust1_cent = init_centroids[0]
clust2_cent = init_centroids[1]
clust3_cent = init_centroids[2]
clust1_cent_CPK = clust1_cent.CPK
clust2_cent_CPK = clust2_cent.CPK
clust3_cent_CPK = clust3_cent.CPK
```

```python
f = lambda x: x['CPK']
for i in range(0,100):
    clust1 = []
    clust2 = []
    clust3 = []
    for i in range(0,df.shape[0]):
        x = y = z = 0
        x = math.sqrt(( df['CPK'][i] - clust1_cent_CPK)**2)
        y = math.sqrt(( df['CPK'][i] - clust2_cent_CPK)**2)
        z = math.sqrt(( df['CPK'][i] - clust3_cent_CPK)**2)
        # print("X: {}, Y: {}, Z:{}".format(x,y,z))
        if x < y and x < z:
            clust1.append(df.iloc[i])
            # print("IF")
            # print(i)
        elif y < x and y < z:
            clust2.append(df.iloc[i])
            # print("ELIF")
            # print(i)
        else:
            clust3.append(df.iloc[i])
            # print("ELSE")
            # print(i)
        # print(i)
    clust1_CPK = [f(x) for x in clust1]
    clust2_CPK = [f(x) for x in clust2]
    clust3_CPK = [f(x) for x in clust3]
    clust1_cent_CPK = statistics.mean(clust1_CPK)
    clust2_cent_CPK = statistics.mean(clust2_CPK)
    clust3_cent_CPK = statistics.mean(clust3_CPK)
# clust1_CPK = [f(x) for x in clust1]
# clust2_CPK = [f(x) for x in clust2]
# clust3_CPK = [f(x) for x in clust3]
# clust1_cent_CPK = statistics.mean(clust1_CPK)
# clust2_cent_CPK = statistics.mean(clust2_CPK)
# clust3_cent_CPK = statistics.mean(clust3_CPK)
```
