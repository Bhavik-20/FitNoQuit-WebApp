```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('exercise_dataset.csv')
df.drop(['Calories per kg'], axis = 1, inplace = True)
df
```

```python
df['CPK'] = 1
for i in range(0, len(df)):
    df['CPK'][i] = (df['130 lb'][i] / 59.1 + df['155 lb'][i] / 70.45 + df['180 lb'][i] / 81.81 + df['205 lb'][i] / 93.18) / 4
df
```

```python
wt1 = []
wt2 = []
wt3 = []
wt4 = []
for i in range(0, len(df['130 lb'])):
    wt1.append(130)
    wt2.append(155)
    wt3.append(180)
    wt4.append(205)
df['weight1'] = wt1
df['weight2'] = wt2
df['weight3'] = wt3
df['weight4'] = wt4

c = []
for i in range(0,len(df['130 lb'])):
    cf = 0
    c1 = df.loc[i][1] - (df['CPK'][i] * 60) 
    c2 = df.loc[i][2] - (df['CPK'][i] * 70) 
    c3 = df.loc[i][3] - (df['CPK'][i] * 82) 
    c4 = df.loc[i][4] - (df['CPK'][i] * 93)
    cf = (c1+c2+c3+c4)/4
    c.append(cf)
c = sum(c) / len(c)
c
```

# Y (calories burnt) = M(CPK) * X(Weight) + C(Constant)

```python
calories = 200
time = [0.5, 0.75, 1] #30 mins, 45 mins, 1hr
w = 70
l = []
for i in range(0, len(df)):
    # cpk = df['CPK'][i] * w
    cpk = df['CPK'][i]*w + c 
    for t in time:
        if cpk * t <= 220 and cpk * t >= 180:
            l.append((df['Activity, Exercise or Sport (1 hour)'][i], t, math.ceil(cpk * t)))
l
```
