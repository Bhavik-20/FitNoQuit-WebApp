```python
gender = input("Enter Gender: ")
age = int(input("Enter age: "))
weight = int(input("Enter weight: "))
height = int(input("Enter height: "))
```

```python
bmr = 0
if gender == 'M':
    bmr = 10 * weight + 6.25 * height - 5 * age + 5
elif gender == 'F':
    bmr = 10 * weight + 6.25 * height - 5 * age - 161
print("BMR =", bmr)
```

```python
print("Choose activity intensity from the given options:\n1. Sedentary (little or no exercise)\n" +
"2. Lightly active (light exercise/sports 1-3 days/week)\n3. Moderately active (moderate exercise/sports 3-5 days/week)\n" +
"4. Very active (hard exercise/sports 6-7 days a week)\n5. Extra active (very hard exercise/sports & a physical job)")
choice = input("Select an option: ")
```

```python
tdee = 0
if choice == '1':
    tdee = bmr * 1.2
elif choice == '2':
    tdee = bmr * 1.375
elif choice == '3':
    tdee = bmr * 1.55
elif choice == '4':
    tdee = bmr * 1.725
elif choice == '5':
    tdee = bmr * 1.9
print("TDEE =", tdee)
```
